### 1.1 Numbers

When we talk about numbers in Python, we generally mean integers, but it could also mean floats.  A float is a decimal number, wheras an integer is a whole number.

````
a = 3
print(a)
print(type(a))

b = 1.5
print(b)
print(type(b))
````

#### Operators
Numbers can be manipulated with several standard mathematical operations.  Note that dividing integers produces an output of a float (decimal number).

````
print(a + 1)  # Addition
print(a - 4)  # Subtraction
print(a * 5)  # Multiplication
print(a / 3)  # Division
print(a ** 2) # Exponentiation
print(a % 2)  # Modulus
````

A very useful operation we might like to do is to increase a number we are storing.  Imagine we are counting something and we want to increment our value by a certain amount, we could do the following:

````
b = 5
print(b)
b = b + 1
print(b)
b = b + 2
print(b)
````

In python, this can be shortened to:

````
b += 1
print(b)
b += 2
print(b)
b += 10
print(b)
````

#### Casting
Some variable types can be represented as different types.  For example an integer can be cast as a float and vice versa. Simply pass the integer into the `float` function.

````
print(float(20))
````

Or the opposite - casting a float to an integer.

````
print(int(20.0))
````

You might have noticed from previous sections that dividing numbers produces a float result (decimal number).

````
a = 5
b = 2
print(a / b)
````

If we want the result to remain an integer (whole number), we can use integer division (//) which casts the result to integer after the division.

````
print(a // b)
````

Or we can achieve the same by casting the float ourselves with the `int` function.

````
print(int(a / b))
````

#### Formatting
Sometimes when printing a float, we want to format it in a specific way - e.g. currencies typically are written to two decimal places. Use the following syntax to format a float, where the `f` stands for float and the `.2` for how many decimal points that decimal should have:

````
d = 2.543
print('This is a formatted number £{:.2f} as a currency'.format(d)) # Print d to two decimal places
````

**Splitting lists**

Python lists can be split up into "sub" lists, for example `[1, 2, 3, 4, 5]` could be split into `[1, 2, 3]` or `[2, 3, 4, 5]` etc.

The notation used `[x:y]`, where `x` is the index of the start of the range, and `y` is the index **after** the last item in the range.

````
characters = ['a', 'b', 'c', 'd', 'e', 'f']
print(characters[0])   # First element of l5
print(characters[0:2]) # First two elements of l5
print(characters[1:3]) # Second and third elements of l5
````

A blank index on the left of `:` indicates the start of the list (i.e. 0). A blank index on the right indicates the end of the list.

````
print(characters[:2])  # Alternative notation for first two elements
print(characters[1:])  # All elements except for the first
print(characters[:])   # All elements of l5
````

Combining this with negative indexes can be very powerful!

````
print(characters[:-1]) # All elements except for the last
print(characters[-2:]) # The last two elements in the list
````

**Zip**

Python has a built-in zip function that pairs up values in multiple lists.

For example, zipping the two lists `[1, 2, 3]` and `['a', 'b', 'c']` will result in `[(1, 'a'), (1, 'b'), (3, 'c')]`.

**Warning**: Make sure the two lists are the same length or you could lose some values!

````
list1 = [1, 2, 3]
list2 = ['a', 'b', 'c']
print(list(zip(list1, list2)))
````

These values pairs are called **tuples**.

can zip together as many lists as you want.

````
list3 = [10, 11, 12]
list4 = [True, True, False]
print(list(zip(list1, list2, list3, list4)))
````

You can zip different sizes of lists, but the number of tuples created will only be as many as the shortest list.

````
list5 = [100, 200, 300, 400]
list(zip(list4, list5))
````

## 2. Comparisons

In many situations, we want to get the program to make a decision based on some condition.  For example, if a number is positive, do one thing, but do another thing if it is negative.  

Python uses the keywords if | elif | else

```
if <some condition>:
    <do one thing>
else:
    <do another thing>
```

**Note**: In Python we don't use curly brackets. Instead we use colons (:) and indentation. Nested code should be four spaces (or one tab character) to the right of its parent.

### 2.1 If/Else

The following is the general syntax structure for an if/else statement in Python:

````
a = 5
if a > 0:
    print('positive')
else:
    print('negative')
````

Sometimes we want to add even more conditions.  For example, if it's raining you should wear waterproof trousers, but if it's sunny you should wear shorts, otherwise you should wear jeans.  Use the `elif` keyword for all of the middle conditions, after the first `if`.

```
if it's raining:
    wear waterproof trousers
elif it's sunny:
    wear shorts
else:
    wear trousers
```

````
a = 5
if a > 0:
    print('positive')
elif a < 0:
    print('negative')
else:
    'zero'
````

### 2.2 Comparison Operators

There are several comparison operators you may need to use for comparing variables to one another.

````
a = 3
b = 5

a == b     # Equality
a > b      # Greater than
a < b      # Less than
a >= b     # Greater than or equal to
a <= b     # Less than or equal to
a != b     # Not equal to
a is b     # Alternative equality
a is not b # Alternative not equal to
````

For booleans, we don't need to check for equality...

````
turned_on = True

if turned_on == True:
    print('Turned on')

if turned_on:
    print('Turned on')
    
if not turned_on:
    print('Turned off')
````

### 2.3 Ternary Operators

Traditionally if/else statements require multiple lines of code. Python has a shorthand version where you can have an if/else in a single line! The syntax is quite simple.

**Remember**: This will only work for an if/else.  Not an if/elif/else.

````
fruit = 'Pear'
print('Yes') if fruit == 'Apple' else print('No')
````

## 3. Loops

### 3.1 For Loops

Loops in Python use the same colon and indent notation as conditional statements.  They follow the form: `for x in iteratable`.

```
for thing in list_of_things:
    do something with thing
```

````
# Loops 10 times
for num in [1,2,3]:
    print(num)
````

````
# Loops 10 times
for i in range(0, 10):
    print(i)
````

**Lists**

We can easily iterate over a list. Traditionally in languages like Java we would do the following:

````
x = ['green', 'eggs', 'and', 'ham']
for i in range(0, len(x)):
    print(x[i])
````

But there is an easier way to do this in Python:

````
x = ['green', 'eggs', 'and', 'ham']
for element in x:
    print(element)
````

**Dictionaries**

Iterating over dictionary elements is also quite easy and there are a couple of options. The first is as follows:

````
shopping_list = {
    'milk': 1.2,
    'eggs': 0.99,
    'bread': 1.39
}
for item in shopping_list:
    # Get the price from the dictionary using the key and format it 
    cost_string = '{:.2f}'.format(shopping_list[item])
    print('The item "%s" costs £%s.'% (item, cost_string))
````

The second option uses the `items` function which gives us direct access to both the key and the value of the given item from the dictionary.

````
shopping_list = {
    'milk': 1.2,
    'eggs': 0.99,
    'bread': 1.39
}
for item, price in shopping_list.items():
    cost_string = '{:.2f}'.format(price)
    print('The item "%s" costs £%s.'% (item, cost_string))
````

Functions/methods allow us to write a piece of reusable code, reducing duplication and making code easier to read and test.

In Python, functions are declared with the word `def`, followed by the function name, parentheses and a colon.

````
def say_hello():
    print('hello')
````

````
say_hello()
say_hello()
say_hello()
````

### 4.2 Parameters

If a function needs some information to perform its operation, it needs parameters.  For example to add two number together, you need to know which two numbers you are adding.  These are defined inside the parentheses like so: 

````
def add(a, b):
    result = a + b
    return result


print(add(5, 6))
````

Just to show you another example of parameters, this time a list printing function.

````
def print_list(their_list):
    for item in their_list:
        print(item)


print_list(['dog', 'cat', 'bird', 'mouse'])
````

### 4.3 Default Parameters

Python has the concept of "default parameters".  When you write a function, you are able to set a default value for one of the parameters.  When the function is called without the parameter specified, the default value will be taken for the execution of the function.  To change the value when calling the function, you must specify the name of the parameter and the value you are assigning to it.

````
def say_hello(loud=False):
    if loud:
        print('HELLO')
    else:
        print('hello')
    return


say_hello()
say_hello(loud=True)
````

### 4.4 Recursion

If you haven't yet come across the concept of recursion, it is essentially when a function references itself, such as in this search algorithm below.

Be warned, recursion can quite easily result in an infinite loop!  This can also be tricky to get your head around so don't worry about this topic.  We will likely not need it in this course.

````
def find_int(the_list, int_to_find):
    if the_list[0] == int_to_find:
        return 0
    return 1 + find_int(the_list[1:], int_to_find)
````

````
print(find_int([1, 2, 3, 4, 5], 4))
````

###4.5 Tuples

Tuples are groups of two or more values. They are unlike lists in that they provide a structure, while lists provide order. For example:

````
val1, val2 = (1, 2)
print(val1)
print(val2)
````

````
val3 = (2, 3)
print(val3[0])
````

This allows us to create functions that return multiple values at once.

````
def multiply_and_add(a, b):
    multiply = a * b
    add = a + b
    return multiply, add

mult, add = multiply_and_add(25, 35)
print(mult)
print(add)
````

### 7.1 Exceptions

Exceptions occur at runtime when the Python interpreter can't run a piece of code for some reason. It isn't always a bad thing, exceptions should be expected and there are ways to easily handle them.

To prevent your program crashing, use a `try:` and `except:` block such as this:

````
exception_test = [1, 2, 3]
try:
    my_val = exception_test[4]
except:
    print('There was an error')
````

Quite often we want to be more specific than this generic error. We can actually access details about the error as below:

````
exception_test = [1, 2, 3]
try:
    my_val = exception_test[4]
except Exception as e:
    print('Error details:')
    print(type(e))
    print(e)
````

Here we are using a generic `Exception` type, which will catch all exceptions. This isn't recommended as there may be different exceptions in different scenarios, which will require different ways to handle the situation.

You can see all the types of exceptions here: https://docs.python.org/2/library/exceptions.html

````
exception_test = [1, 2, 3]
try:
    my_val = exception_test[4]
except IndexError:
    print('Handling index error')
except IOError:
    print('Handling IO error')
except Error:
    print('Handling all other errors')
````

It's also possiblle in Python to execute code only if there were no exceptions using an `else:` statement:

````
try:
    print('Testing')
except:
    print('There was an error')
else:
    print('There was no error')
print('This will execute either way')
````

**Debugging**

If you ever need to use a debugger in a Python notebook, it is a little bit awkward, but `Pdb` is the way to go.

`debug.set_trace()` will allow you to create a breakpoint in your code.  A breakpoint is a place that will pause the execution of the program so that you as the programmer can do some investigating (usually to fig bugs).

````
from IPython.core.debugger import Pdb
debug = Pdb()
def add(var1, var2):
    debug.set_trace() # Create a breakpoint
    result = var1 + var2
    return result
````

### 7.3 Data Visualisation

When it comes to machine learning, we're going to be doing quite a lot of visualisation of data, to help us get a better understanding of what's going on. Matplotlib is a graphing library that we will use to plot graphs from data.

Below is a simple example of plotting `[2, 4, 5, 8]` (y-axis) against `[1, 2, 3, 4]` (x-axis).

````
# For notebooks, we have to include this line to allow graphs to display in the notebook
%matplotlib inline

import matplotlib.pyplot as plt
plt.plot([1, 2, 3, 4], [2, 4, 5, 8])
plt.ylabel('y-axis')
plt.xlabel('x-axis')
plt.show()
````

Let's try this out on something more meaningful. Using the CSV file from before, we want to try and see how the values have changed over time.

The problem is, our time column (Quarter Year) isn't numeric. We need to clean this up so that it's something like `2005.25` rather than `Q2 2005`. We can use the `apply` function to do this.

Note that this is an example of Feature Engineering!

````
def clean_year(row):
    components = row.split(' ') # e.g. splits 'Q1 2005' into ['Q1', '2005']
    year = int(components[1]) # Converts year from a string to an int
    quarter = int(components[0][1]) # Removes the 'Q' from the quarter (e.g. Q1 -> 1)
    decimal_quarter = (quarter - 1) * 0.25 # Q1 -> 0.0, Q2 -> 0.25 etc
    return str(year + decimal_quarter)

data['clean_year'] = data['Quarter Year'].apply(clean_year)
````

Now we can nicely plot the Price Index against time.

````
plt.plot(data['clean_year'], data['NI Residential Property Price Index'])
plt.ylabel('NI Residential Property Price Index')
plt.xlabel('Year')
plt.show()
````

It's also possible to plot multiple lines on a single graph, and provide a legend as below:

````
plt.plot(data['clean_year'], data['NI Detached Property Price Index'])
plt.plot(data['clean_year'], data['NI SemiDetached Property Price Index'])
plt.plot(data['clean_year'], data['NI Terrace Property Price Index'])
plt.plot(data['clean_year'], data['NI Apartment Price Index'])
plt.plot(data['clean_year'], data['NI Residential Property Price Index'])

plt.legend([
        'Detached',
        'SemiDetached',
        'Terrace',
        'Apartment',
        'Residential'
    ])
plt.xlabel('Year')

plt.show()
````
