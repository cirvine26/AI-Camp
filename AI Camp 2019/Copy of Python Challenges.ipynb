# 2. Convert string to appropriate type

Write a function that converts a string to either an int or a float, depending on if it has a decimal point. It should return 0 for invalid input.

E.g.

    convert('1.5') -> 1.5
    convert('780') -> 780

```python
def convert(string):
  if "." in string:
    return(float(string))
  else:
    try: 
      return(int(string))
    except:
      return(0)
```

```python
# Convert tests
if (convert('1.5') == 1.5 and
    convert('780') == 780 and
    convert('pupper') == 0 and
    convert('-28.07') == -28.07):
    print('Success')
else:
    print('Fail')
```

# 3. What can I afford?

Write a function that takes a dictionary of shopping items with their associated cost, and a float (your budget), and return a dictionary with the items you can afford with the given budget. Maximise the number of items to be bought.

```python
import numpy as np
def afford(shopping_items, budget):
    
    #Initialsing neede values 
    values = shopping_items.values()
    values=np.array(values)
    total_price = np.sum(values)
    
    #Ordering the dictionary into a list of tuples 
    sorted_prices = sorted(shopping_items.iteritems(), key = lambda x : x[1])
    
    #Initialising for loop to identify if price is within budget
    while budget <= total_price:
      del sorted_prices[-1]
      values = np.array([x[1] for x in sorted_prices])
      total_price = np.sum(values)
      
    final_dictionary = dict(sorted_prices)
    return final_dictionary      
```

```python
# Afford tests
afford1 = {
    'ketchup': 1.69,
    'crisps multipack': 3.50,
    'beer': 3.99,
    'bread': 0.99,
    'salad': 1.20,
    'beans': 0.8
}
afford1_result = {
    'ketchup': 1.69,
    'crisps multipack': 3.50,
    'bread': 0.99,
    'salad': 1.20,
    'beans': 0.8
}
budget1 = 10.0

afford2 = {
    'birthday card': 2.99,
    'makeup': 8.00,
    'charms': 75.00
}
afford2_result = {
    'birthday card': 2.99,
    'makeup': 8.00,
    'charms': 75.00
}
budget2 = 99.99

afford3 = {
    'tv': 349.0,
    'ps4': 280.0,
    'sound bar': 99.0,
    'games': 100.0,
    'accessories': 49.0
}
afford3_result = {
    'sound bar': 99.0,
    'games': 100.0,
    'accessories': 49.0
}
budget3 = 500.0

if (afford(afford1, budget1) == afford1_result and
    afford(afford2, budget2) == afford2_result and
    afford(afford3, budget3) == afford3_result):
    print 'Success'
else:
    print 'Fail'
```

# 4. Sin and Cos graph

Create a graph using matplotlib that plots both a Sin and Cos wave from x = 0 to x = 10. It should look something like this:

![Sin and Cos plotted](https://ai-camp-content.s3.amazonaws.com/waves.png)

Note: Numpy might come in handy!


```python
import numpy as np 
import matplotlib.pyplot as plt

x = np.arange(0,11,0.01)
y_sin=np.sin(x)
y_cos=np.cos(x)

y_ticks = np.arange(-1,1.5,0.5)

plt.plot(x,y_sin, color = 'b')
plt.plot(x,y_cos, color = 'g')
plt.yticks(y_ticks)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
```
